```python
# ==============================================
# ğŸ“Š Add Missing Features to Dataset (FIXED)
# ==============================================
import pandas as pd
import numpy as np

print("="*60)
print("ğŸ”¥ Loading Dataset...")
print("="*60)

# Load dataset
df = pd.read_csv("NVDA_alpha.csv")  # Using your original file
print(f"\nğŸ“‹ Original Dataset Shape: {df.shape}")
print(f"ğŸ“‹ Current Columns: {df.columns.tolist()}\n")

# ==========================================
# 1ï¸âƒ£ Add ALL Required Features
# ==========================================
print("ğŸ”§ Adding Missing Features...\n")

# 1. SMA5 (Simple Moving Average - 5 days)
print("   âœ… Adding SMA5...")
df['SMA5'] = df['Close'].rolling(window=5).mean()

# 2. SMA20 (using existing SMA_20 or creating new)
if 'SMA_20' in df.columns:
    print("   âœ… Renaming SMA_20 to SMA20...")
    df['SMA20'] = df['SMA_20']
else:
    print("   âœ… Adding SMA20...")
    df['SMA20'] = df['Close'].rolling(window=20).mean()

# 3. SMA50 (Simple Moving Average - 50 days)
print("   âœ… Adding SMA50...")
df['SMA50'] = df['Close'].rolling(window=50).mean()

# 4. Price_Change (already exists, verify it's correct)
if 'Price_Change' not in df.columns:
    print("   âœ… Adding Price_Change...")
    df['Price_Change'] = ((df['Close'] - df['Open']) / df['Open']) * 100
else:
    print("   â„¹ï¸ Price_Change already exists")

# 5. RSI (already exists, keep it)
print("   â„¹ï¸ RSI already exists")

# 6. MACD (already exists, keep it)
print("   â„¹ï¸ MACD already exists")

# 7. MACD_Signal (NEW - this is likely missing)
print("   âœ… Calculating MACD_Signal...")
df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

# 8. Direction (already exists, keep it)
if 'Direction' not in df.columns:
    print("   âœ… Adding Direction...")
    df['Direction'] = np.where(df['Price_Change'] > 0, 1, 0)
else:
    print("   â„¹ï¸ Direction already exists")

print("\n" + "="*60)
print("âœ… All Features Added Successfully!")
print("="*60)

# ==========================================
# 2ï¸âƒ£ Remove NaN values
# ==========================================
print(f"\nğŸ“Š Handling NaN values...")
print(f"   Before: {df.shape[0]} rows")
rows_before = df.shape[0]
df_clean = df.dropna()
print(f"   After: {df_clean.shape[0]} rows")
print(f"   Removed: {rows_before - df_clean.shape[0]} rows\n")

# ==========================================
# 3ï¸âƒ£ Select and Reorder Columns
# ==========================================
print("ğŸ“‹ Selecting Final Columns...")

# Keep only the columns we need
final_columns = [
    'date', 'Open', 'High', 'Low', 'Close', 'Volume',
    'SMA5', 'SMA20', 'SMA50', 
    'RSI', 'MACD', 'MACD_Signal',
    'BB_upper', 'BB_middle', 'BB_lower',
    'Price_Change', 'Direction'
]

# Create final dataframe with selected columns
df_final = df_clean[final_columns].copy()

print(f"âœ… Final columns selected: {len(final_columns)}\n")

# ==========================================
# 4ï¸âƒ£ Display Dataset Info
# ==========================================
print("="*60)
print("ğŸ“‹ Final Dataset Information")
print("="*60)
print(f"\nâœ… Final Shape: {df_final.shape}")
print(f"âœ… Total Columns: {len(df_final.columns)}")
print(f"\nğŸ“Œ Columns in Dataset:")
for i, col in enumerate(df_final.columns, 1):
    print(f"   {i}. {col}")

print(f"\nğŸ“Š Dataset Preview (first 5 rows):")
print(df_final.head())

print(f"\nğŸ“Š Dataset Statistics:")
print(df_final.describe())

# ==========================================
# 5ï¸âƒ£ Verify Required Features
# ==========================================
print("\n" + "="*60)
print("ğŸ¯ Checking Required Features")
print("="*60)

required_features = ['Close', 'SMA5', 'SMA20', 'SMA50', 
                     'Price_Change', 'RSI', 'MACD', 'MACD_Signal']

print("\nRequired features for model training:")
all_present = True
for feature in required_features:
    if feature in df_final.columns:
        print(f"   âœ… {feature}")
    else:
        print(f"   âŒ {feature} - MISSING!")
        all_present = False

if all_present:
    print("\nğŸ‰ All required features are present!")
else:
    print("\nâš ï¸ Some features are missing!")

# ==========================================
# 6ï¸âƒ£ Save Updated Dataset
# ==========================================
print("\n" + "="*60)
print("ğŸ’¾ Saving Updated Dataset")
print("="*60)

output_file = "NVDA_dataset_updated.csv"
df_final.to_csv(output_file, index=False)
print(f"\nâœ… Dataset saved as: {output_file}")
print(f"âœ… Rows: {df_final.shape[0]}")
print(f"âœ… Columns: {df_final.shape[1]}")
print(f"âœ… Ready for model training!\n")

# ==========================================
# 7ï¸âƒ£ Verify the saved file
# ==========================================
print("ğŸ” Verifying saved file...")
df_verify = pd.read_csv(output_file)
print(f"âœ… Loaded back successfully!")
print(f"âœ… Shape: {df_verify.shape}")
print(f"âœ… Columns: {df_verify.columns.tolist()}\n")

print("="*60)
print("âœ¨ Feature Engineering Complete!")
print("="*60)
print("\nğŸ“ Next Steps:")
print("   1. File 'NVDA_dataset_updated.csv' is ready")
print("   2. Update your training script to use this file")
print("   3. Run: python train_model.py\n")

# ==========================================
# 8ï¸âƒ£ Quick Data Quality Check
# ==========================================
print("="*60)
print("ğŸ” Data Quality Check")
print("="*60)
print(f"\nâœ… No missing values: {df_final.isnull().sum().sum() == 0}")
print(f"âœ… Date range: {df_final['date'].min()} to {df_final['date'].max()}")
print(f"âœ… Total trading days: {df_final.shape[0]}")
print(f"\nâœ… Feature ranges:")
for col in ['SMA5', 'SMA20', 'SMA50', 'RSI', 'MACD']:
    print(f"   {col}: {df_final[col].min():.2f} to {df_final[col].max():.2f}")
```
